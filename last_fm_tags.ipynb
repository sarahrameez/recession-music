```python
billboard_tags['tags_api'] = None
```

```python
for index, row in billboard_tags.iloc[2000:].iterrows():
        if pd.isna(row['tags']):
            tags_to_add = ''
            track = row['title']
            artist = row['performer']
            result = network.get_track(track, artist)
            try:
                tag_list = []
                for tag in result.get_top_tags():
                    tag_list.append(tag[0].name)
                if len(tag_list) > 0:
                    tags_to_add = ', '.join(tag_list)
                    print(index)
            except:
                continue
            billboard_tags.loc[index, 'tags_api'] = tags_to_add if tags_to_add != "" else None
```

```python
fill_1 = len(billboard_tags) - billboard_tags['tags_api'].isna().sum()
fill_2 = len(billboard_tags) - billboard_tags['tags'].isna().sum()

print((fill_1)/len(billboard_tags))
print((fill_2)/len(billboard_tags))
print((fill_1 + fill_2)/len(billboard_tags))
```
